```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
### Flan T5 XXL
# Run this cell twice for proper font-size formatting
x = ['USA','AUS','UK','SA','SG']
y1 = [0.2866360643,	0.2342404296,	0.2734696881,	0.2862101706,	0.2819161023]
y2 = [0.3347177035,	0.2714131537,	0.2958907768,	0.3540878821,	0.3095167055]

x_pos = np.arange(len(x))
width = 0.4

plt.bar(x_pos - width/2, y1, width, label='Y1', color="#917A70", align='center')
plt.bar(x_pos + width/2, y2, width, label='Y2', color="#75ca9a", align='center')

plt.xlabel('Countries', labelpad=15)
plt.ylabel('IAA', labelpad=15)
plt.ylim(0,0.6)
plt.xticks(x_pos, x, rotation=0)

plt.legend(["w/o country info","w/ country info"])
plt.rcParams.update({'font.size': 21})
#plt.show()
plt.savefig('GC/GC_C_T5.pdf', format='pdf', bbox_inches='tight')
```
