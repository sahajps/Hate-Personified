```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
# Lang Info IAA

x = ['Ar','Fr','De','Hi']
y1 = [0.58, 0.3438208375, 0.5015369278, 0.2420435511]
y2 = [0.66, 0.4251040688, 0.536508842, 0.3714021041]

x_pos = np.arange(len(x))
width = 0.4

plt.bar(x_pos - width/2, y1, width, label='Y1', color='#ffa600', align='center')
plt.bar(x_pos + width/2, y2, width, label='Y2', color='#bc5090', align='center')

plt.xlabel('Languages', labelpad=15)
plt.ylabel('IAA', labelpad=15)
plt.ylim(0,1)
plt.xticks(x_pos, x, rotation=0)
plt.legend(["w/o language info","w/ language info"])
plt.rcParams.update({'font.size': 21})
#plt.show()
plt.savefig('ML/ML_LI.pdf', format='pdf', bbox_inches='tight')
```
